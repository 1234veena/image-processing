```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('cat_damaged.png')
plt.title('damaged img')
plt.imshow(img)
plt.show()

mask = cv2.imread('cat_mask.png', 0)

dst = cv2.inpaint(img, mask, 3, cv2.INPAINT_NS)
plt.title('restored img')
plt.imshow(dst)
plt.show()
```
