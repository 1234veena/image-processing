```python
import glob
import cv2
import matplotlib.pyplot as plt
path='animals/*.*'
images=[cv2.imread(images)for images in glob.glob(path)]
for i in range(len(images)):
    cv2.imshow("original",images[i])
    cv2.waitKey(0)
    Gaussian = cv2.GaussianBlur(images[i], (9, 9), 0)
    cv2.imshow(" Gaussian", Gaussian)
    cv2.waitKey(0)
    median = cv2.medianBlur(images[i], 5)
    cv2.imshow("median", median)
    cv2.waitKey(0)
    bilateral = cv2.bilateralFilter(images[i], 9, 75, 75)
    cv2.imshow("bilateral",bilateral)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import glob
import cv2
import matplotlib.pyplot as plt
path='animals/*.*'
images=[cv2.imread(images)for images in glob.glob(path)]
for i in range(len(images)):
    plt.imshow(images[i])
    plt.axis("off")
    plt.title("original")
    plt.show()
    Gaussian = cv2.GaussianBlur(images[i], (9, 9), 0)
    plt.imshow(Gaussian)
    plt.axis("off")
    plt.title("gaussian")
    plt.show()
    median = cv2.medianBlur(images[i], 5)
    plt.imshow(median)
    plt.axis("off")
    plt.title("median")
    plt.show()
    bilateral = cv2.bilateralFilter(images[i], 9, 75, 75)
    plt.imshow(bilateral)
    plt.axis("off")
    plt.title("bilateral")
    plt.show()
```
