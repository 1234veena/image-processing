```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img =cv2.imread('nature.jpg')
cv2.line(img, (200, 600), (100, 160), (0, 0, 255), 5)
cv2.rectangle(img,(700,800), (200,300),(0,255,255),10)
cv2.circle(img, (200,500), 65, (255,0,0), -1)
plt.title('img')
plt.imshow(img)
plt.show()
```
