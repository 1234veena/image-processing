```python
# importing the library
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open("photo.jpg")
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# text Watermark
from PIL import ImageFont
from PIL import ImageDraw
watermark_image = image.copy()

draw = ImageDraw.Draw(watermark_image)

w, h = image.size
x, y = int(w / 2), int(h / 2)
if x > y:
    font_size = y
elif y > x:
    font_size = x
else:
    font_size = x

font = ImageFont.truetype("arial.ttf", int(font_size/2))

# add watermark
draw.text((x, y), "camera", fill=(0, 0, 0), font=font, anchor='ms')
plt.subplot(1, 2, 1)
plt.title("black text")
plt.axis('off')
plt.imshow(watermark_image)

# add watermark
draw.text((x, y), "camera", fill=(255, 255, 255), font=font, anchor='ms')
plt.subplot(1, 2, 2)
plt.title("white text")
plt.axis('off')
plt.imshow(watermark_image)
```

```python
# image watermark
size = (600, 200)
crop_image = image.copy()
crop_image.thumbnail(size)

# add watermark
copied_image = image.copy()
copied_image.paste(crop_image, (500, 200))
plt.imshow(copied_image)
plt.show()
```
